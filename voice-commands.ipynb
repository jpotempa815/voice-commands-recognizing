```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import scipy
import pandas as pd
```

```python
path = r"C:\Users\julia\Desktop\studia\semestr 9\Uczenie maszynowe\voice-commands-recognizing\sounds"

words = ["yes", 'no', 'up', 'down', 'left', 'right', 'on', 'off', 'cake', 'wave']
num = 72
sac_signals = []
sample_rates = []
idx = []

for word in words:
    for i in range(1, num):
        if word == "cake" or word == "wave" and i >= 37:
            continue
        sac_file = f"{path}\{word}_{i}.wav"
        sac_signal, sample_rate = librosa.load(sac_file)
        sac_signals.append(sac_signal)
        sample_rates.append(sample_rate)
        idx.append(word)
df = pd.DataFrame(data={'signal': sac_signals, 'sample_rate': sample_rates, 'word' : idx})
df
```

```python
fourier_transforms = []
for i, x in df.iterrows():
    fourier_transforms.append(np.fft.fft(x['signal']))
df['fourier_transform'] = fourier_transforms
df
```
